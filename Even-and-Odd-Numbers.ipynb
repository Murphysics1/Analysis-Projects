## What Percentage Of Numbers Are Even?

So it turns out that there is no formal proof that 1/2 of all numbers are even. By doing a 1:1 comparison, you could make a successful argument that any ratio is correct

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

Using the following two ideas, we can create a list of ratios for the number of even and odd numbers to all numbers. Intuition would tell us that at infinity, the ratio is 50% for both even and odd 
$$ \text{even ratio} = \frac{\sum_{i=0}^{n} i \text{ mod(2)}}{\sum_{i=0}^{n} 1} $$ <br><br> $$ \text{odd ratio} = \frac{\sum_{i=0}^{n} (i+1) \text{ mod(2)}}{\sum_{i=0}^{n} 1} $$ <br>

Create functions that will count the number of even numbers, odd numbers, and numbers checked

```python
def even(lim):
    
    s = 0
    
    for num in range(1,lim+2):
        s = s+ num % 2
        #even_data.append(s)
    return s
```

```python
def odd(lim):
    
    s = 0
    
    for num in range(1,lim+2):
        s = s+ (num+1) % 2
    return s
```

```python
def one(lim):
    
    s = 0
    
    for num in range(1,lim+2):
        s += 1
    
    return s
```

Create the data sets by dividing the rolling sums of evens and odds to totals

```python
lim = 150
x = [num for num in range(lim)]
even_ratio = [even(num)/one(num) for num in range(lim)]
odd_ratio = [odd(num)/one(num) for num in range(lim)]
```

Generate Graphs

```python
plt.rcParams["figure.figsize"] = (15,9)
plt.plot(x,odd_ratio)
plt.plot(x,even_ratio)
plt.xlabel("Test Number",fontsize = 15)
plt.ylabel("Ratio",fontsize = 15)
plt.title("Ratio of Even and Odd Numbers",fontsize = 20)
plt.legend(['Odd','Even'],fontsize = 15,loc = 4)
plt.show()
```

Intuitively, it seems that the limit is being squeezed to 50% even and 50% odd. If we could find the evelope equation and find its limit at infinity, then that would be convincing enough for me. What is interesting the note is that no matter what the x value is, the sum of the even ratio to odd ratio is always 1. This is good evidence that the model is working
