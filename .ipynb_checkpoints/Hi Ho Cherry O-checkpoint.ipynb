```python
import matplotlib.pyplot as plt
from random import randint
import seaborn as sns
import pandas as pd
sns.set()
```

```python
score = {"chicken":-2,"dog":-2,"basket":"all","1":1,"2":2,"3":3,"4":4}
spin = ["chicken",'dog','basket',"1","2","3","4"]
```

What a typical game might look like

```python
count = 0
turns = []

while count < 10:
    #generate the turn
    cycle = score[spin[randint(0,6)]]
    
    #if you land on the basket
    if cycle == "all":
        count = 0
    else:
        count += cycle
        #set a floor at 0
        if count <0:
            count = 0
        #set a ceiling at 10
        if count >10:
            count = 10
    turns.append(count)
    
print(turns)
print(f'It took {len(turns)} turns to reach 10')
plt.plot(turns)
plt.xlabel("Turn Number")
plt.ylabel("Curent Score")
plt.title('Hi Ho Cherry O')
plt.show()
```

What are the stats for 1,000,000 games

```python
def game():
    
    count = 0
    turns = 0

    while count < 10:
        cycle = score[spin[randint(0,6)]]
        if cycle == "all":
            count = 0
        else:
            count += cycle
            if count <0:
                count = 0
        turns += 1
    
    return turns
```

```python
#Simulate a lot of games

def simulation(n):
    
    results = [game() for i in range(n)]

    return pd.Series(results)
```

```python
data = simulation(1000000)
```

```python
sns.histplot(data,bins=70,binrange=(0,70),kde=True)
plt.ylabel("Games Won")
plt.xlabel("Turn Number")
plt.title('Hi Ho Cherry O',size =15)
plt.xlim([0,75])
plt.show()
```

```python
data.describe()
```

The fact that my kid wins quickly is statistically not surprising. 

Next, we'll bootstrap our data to see if there would be a wide shift in the mean through repeated experimentation. 

Bootstrap Distribution

```python
def bootstrap(n,b=None):
    
    #run simulation
    raw_data = pd.Series([game() for i in range(n)])
    
    #adjust the data set for a mean of zero
    mu = raw_data.mean()
    data2 = pd.Series([(raw_data[i]-mu) for i in range(n)])
    
    #If sampling less than the number of data points
    x = b
    if b == None:
        x = n
    
    #Create a distribution of means
    bdist = []
    for i in range(n):
        s = pd.Series([data2[randint(0,n-1)] for i in range(x)])
        bdist.append(s.mean())
    
    #print('Mean is {:.0f}'.format(raw_data.mean()))
    #print('Adjusted Mean is {:.0f}'.format(data2.mean()))
    #print('Boot Mean is {:.0f}'.format(bdist.mean()))
    
    return raw_data,data2,pd.Series(bdist)
    
```

```python
raw_data,adjusted_data,bts_data = bootstrap(10000)
```

```python
mu = raw_data.mean()

fig, axes = plt.subplots(1, 3, figsize=(15, 5), sharey=True)

sns.histplot(ax=axes[0],data=raw_data,bins=70,binrange=(0,70),kde=True)
sns.histplot(ax=axes[1],data=adjusted_data,bins=70,binrange=(0-mu,70-mu),kde=True)
sns.histplot(ax=axes[2],data=bts_data,bins=70,kde=True)
fig.suptitle('Hi Ho Cherry O')
axes[0].set(title='Raw Data',xlabel='Turn Number')

axes[1].set(title='Adjusted')
axes[2].set(title='Bootstrapped Distribution',xlabel = 'Deviation from the Mean')
plt.show()
```

The bootstrap distribution appears normal with a very small range. It seems that a mean of 15.8 is very consistant
